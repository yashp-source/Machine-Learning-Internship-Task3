# Data cleaning

```python
data=iri.values

X=data[:,0:4]
Y=data[:,4]
print(Y)
```

```python
print(X)
```

```python
#Here, extracting the features into X and the target variable (species) into Y.
```

# Training data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
print(X_train)
```

# Testing data

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
# Here splitting the data into training and testing sets using the train_test_split function.
```

# LogisticRegression

```python

from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
# Here i am creating a Logistic Regression model and training it on the training data.
```

# Prediction

```python
prediction1= model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
for i in range (len(prediction1)):
    print(y_test[i],prediction1[i])
```
